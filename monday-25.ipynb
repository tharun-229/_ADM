```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset from CSV file
file_path = "/content/titanic.csv"
df = pd.read_csv(file_path)

def removal_box_plot(df, column):
    # Original box plot
    sns.boxplot(x=df[column])
    plt.title(f'Original Box Plot of {column}')
    plt.show()

    # Compute IQR for dynamic thresholding
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Removing outliers
    df_cleaned = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

    # Box plot after removing outliers
    sns.boxplot(x=df_cleaned[column])
    plt.title(f'Box Plot without Outliers of {column}')
    plt.show()

    return df_cleaned

# Apply the function to 'Age' column (Change if needed)
no_outliers = removal_box_plot(df, 'Age')

# Display first few rows after outlier removal
print(no_outliers.head())

```
